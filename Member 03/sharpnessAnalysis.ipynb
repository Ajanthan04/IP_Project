```python
import cv2
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

mc_input = r"C:\Users\Imashi Rajapaksha\Desktop\FRAME_SELECTION\FRAME_SELECTION_MC"
gc_input = r"C:\Users\Imashi Rajapaksha\Desktop\FRAME_SELECTION\FRAME_SELECTION_GC"

mc_output = r"C:\Users\Imashi Rajapaksha\Desktop\FRAME_SELECTION\SHARP_MC"
gc_output = r"C:\Users\Imashi Rajapaksha\Desktop\FRAME_SELECTION\SHARP_GC"

comparison_folder = r"C:\Users\Imashi Rajapaksha\Desktop\FRAME_SELECTION\COMPARISON_RESULTS"

os.makedirs(mc_output, exist_ok=True)
os.makedirs(gc_output, exist_ok=True)
os.makedirs(comparison_folder, exist_ok=True)

kernel = np.array([[0,-1,0],
                   [-1,5,-1],
                   [0,-1,0]])

results = []

def sharpness_value(image):
    return cv2.Laplacian(image, cv2.CV_64F).var()

def process_images(input_folder, output_folder, label):

    for file in os.listdir(input_folder):

        if file.endswith(".jpg") or file.endswith(".png"):

            path = os.path.join(input_folder,file)

            img = cv2.imread(path)
            gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

            gaussian = cv2.GaussianBlur(gray,(5,5),0)

            sharpened = cv2.filter2D(gaussian,-1,kernel)

            original_sharp = sharpness_value(gray)
            sharpen_sharp = sharpness_value(sharpened)

            cv2.imwrite(os.path.join(output_folder,file),sharpened)

            plt.figure(figsize=(8,4))

            plt.subplot(1,2,1)
            plt.imshow(gray,cmap='gray')
            plt.title(f"Original\nSharpness={original_sharp:.2f}")
            plt.axis("off")

            plt.subplot(1,2,2)
            plt.imshow(sharpened,cmap='gray')
            plt.title(f"Sharpened\nSharpness={sharpen_sharp:.2f}")
            plt.axis("off")

            save_name = label + "_" + file
            plt.savefig(os.path.join(comparison_folder,save_name))
            plt.close()

            results.append([file,label,original_sharp,sharpen_sharp])

            print("Processed:",file)

process_images(mc_input, mc_output, "MC")
process_images(gc_input, gc_output, "GC")

df = pd.DataFrame(results,columns=["Frame","Method","Original Sharpness","Sharpened Sharpness"])
df.to_csv(r"C:\Users\Imashi Rajapaksha\Desktop\FRAME_SELECTION\sharpness_values.csv",index=False)

print("Processing complete!")
```
