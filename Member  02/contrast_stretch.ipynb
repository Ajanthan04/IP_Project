```python
import cv2
import os
import numpy as np

input_folders = {
    "gaussian": "denoised_gaussian",
    "median": "denoised_median"
}

output_folders = {
    "gaussian": "enhanced_gaussian_contrast",
    "median": "enhanced_median_contrast"
}

# Create output folders
for folder in output_folders.values():
    os.makedirs(folder, exist_ok=True)

def contrast_stretch(img):
    min_val = np.min(img)
    max_val = np.max(img)
    stretched = ((img - min_val) / (max_val - min_val) * 255).astype(np.uint8)
    return stretched

for method, folder in input_folders.items():
    print(f"\nProcessing Contrast Stretching: {method.upper()}")
    for filename in sorted(os.listdir(folder)):
        if filename.endswith(".png"):
            img = cv2.imread(os.path.join(folder, filename), cv2.IMREAD_GRAYSCALE)
            if img is None:
                continue
            stretched = contrast_stretch(img)
            cv2.imwrite(os.path.join(output_folders[method], filename), stretched)

print("\nContrast Stretching Completed!")
```
