```python
import pandas as pd
import matplotlib.pyplot as plt
import os

csv_files = {
    "gaussian": "contrast_values_gaussian.csv",
    "median": "contrast_values_median.csv"
}

plot_folder = "contrast_plots"
os.makedirs(plot_folder, exist_ok=True)

for method, csv_file in csv_files.items():
    df = pd.read_csv(csv_file)

    frames = df["Frame"]
    orig_contrast = df["Original Contrast"]
    clahe_contrast = df["CLAHE Contrast"]

    plt.figure(figsize=(12,6))
    plt.plot(frames, orig_contrast, label="Contrast Stretched", marker='o')
    plt.plot(frames, clahe_contrast, label="CLAHE Enhanced", marker='o')
    plt.xticks(rotation=90)
    plt.xlabel("Frame")
    plt.ylabel("Contrast (Std Dev)")
    plt.title(f"Contrast Comparison - {method.capitalize()} Frames")
    plt.legend()
    plt.tight_layout()

    plot_path = os.path.join(plot_folder, f"contrast_plot_{method}.png")
    plt.savefig(plot_path)
    plt.close()
    print(f"{method.capitalize()} contrast plot saved: {plot_path}")

print("\nAll contrast plots generated successfully!")
```
