```python
import cv2
import os
import numpy as np
import pandas as pd

original_folders = {
    "gaussian": "enhanced_gaussian_contrast",
    "median": "enhanced_median_contrast"
}

clahe_folders = {
    "gaussian": "enhanced_gaussian_clahe",
    "median": "enhanced_median_clahe"
}

comparison_folders = {
    "gaussian": "comparison_images_gaussian",
    "median": "comparison_images_median"
}

for folder in comparison_folders.values():
    os.makedirs(folder, exist_ok=True)

for method in original_folders.keys():
    orig_folder = original_folders[method]
    clahe_folder = clahe_folders[method]
    comparison_folder = comparison_folders[method]

    contrast_data = []

    print(f"\nProcessing {method.upper()} frames...")

    for filename in sorted(os.listdir(orig_folder)):
        if not filename.endswith(".png"):
            continue

        orig_path = os.path.join(orig_folder, filename)
        clahe_path = os.path.join(clahe_folder, filename)

        orig_img = cv2.imread(orig_path, cv2.IMREAD_GRAYSCALE)
        clahe_img = cv2.imread(clahe_path, cv2.IMREAD_GRAYSCALE)
        if orig_img is None or clahe_img is None:
            continue

        # Compute contrast (standard deviation)
        orig_contrast = np.std(orig_img)
        clahe_contrast = np.std(clahe_img)
        contrast_data.append([filename, orig_contrast, clahe_contrast])


        comparison_img = np.hstack((orig_img, clahe_img))

        save_path = os.path.join(comparison_folder, filename)
        cv2.imwrite(save_path, comparison_img)

    csv_path = f"contrast_values_{method}.csv"
    contrast_df = pd.DataFrame(contrast_data, columns=["Frame","Original Contrast","CLAHE Contrast"])
    contrast_df.to_csv(csv_path, index=False)
    print(f"{method.upper()} comparison done. Images in '{comparison_folder}', contrast values in '{csv_path}'")

print("\nAll comparisons completed successfully!")
```
