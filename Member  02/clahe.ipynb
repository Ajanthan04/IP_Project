```python
import cv2
import os
import numpy as np

# Input folders: use the contrast-stretched frames as input
input_folders = {
    "gaussian": "enhanced_gaussian_contrast",
    "median": "enhanced_median_contrast"
}

# Output folders: CLAHE enhanced frames
output_folders = {
    "gaussian": "enhanced_gaussian_clahe",
    "median": "enhanced_median_clahe"
}

# Create output folders if they don't exist
for folder in output_folders.values():
    os.makedirs(folder, exist_ok=True)

# Function: Apply optimized CLAHE
def apply_optimized_clahe(img, clip_limit=2.0, tile_grid_size=(16,16), blend_alpha=0.8):
    """
    Apply CLAHE with optional blending to reduce roughness.
    - img: input grayscale image
    - clip_limit: CLAHE clip limit
    - tile_grid_size: size of tiles
    - blend_alpha: weight of CLAHE output when blending with original
    """
    clahe = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=tile_grid_size)
    clahe_img = clahe.apply(img)

    # Optional smoothing to reduce roughness
    clahe_img = cv2.GaussianBlur(clahe_img, (3,3), 0)

    # Blend with original to preserve natural look
    final_img = cv2.addWeighted(clahe_img, blend_alpha, img, 1 - blend_alpha, 0)
    return final_img

# Process frames
for method, folder in input_folders.items():
    print(f"\nApplying Optimized CLAHE on {method.upper()} frames...")
    for filename in sorted(os.listdir(folder)):
        if not filename.endswith(".png"):
            continue

        img_path = os.path.join(folder, filename)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            continue

        enhanced_img = apply_optimized_clahe(img, clip_limit=2.0, tile_grid_size=(16,16), blend_alpha=0.8)

        # Save output
        save_path = os.path.join(output_folders[method], filename)
        cv2.imwrite(save_path, enhanced_img)

        # Print contrast for report
        orig_contrast = np.std(img)
        new_contrast = np.std(enhanced_img)
        print(f"{filename}: Original Contrast={orig_contrast:.2f}, CLAHE Contrast={new_contrast:.2f}")

print("\nOptimized CLAHE Completed! Enhanced frames saved successfully.")
```
