```python
import pandas as pd

# --- Load the CSV file generated from your PSNR/MSE code ---
csv_file = "denoising_metrics.csv"
df = pd.read_csv(csv_file)

# --- Choose metric to sort by ---
# We pick Gaussian PSNR for ranking; you can change to Median if you want
sorted_df = df.sort_values(by="PSNR_Gaussian", ascending=False)

# --- Select top 3 frames ---
top_frames = sorted_df.head(3)

# --- Print the results ---
print("Top 3 frames with highest PSNR (best denoising):\n")
for idx, row in top_frames.iterrows():
    print(f"Frame {row['Frame']}:")
    print(f"  PSNR Gaussian = {row['PSNR_Gaussian']}")
    print(f"  PSNR Median   = {row['PSNR_Median']}")
    print(f"  MSE Gaussian = {row['MSE_Gaussian']}")
    print(f"  MSE Median   = {row['MSE_Median']}")
    print("-"*40)
```
