```python
import cv2
import os
import numpy as np

input_folder = "extracted_frames"
output_gaussian = "denoised_gaussian"
output_median = "denoised_median"

os.makedirs(output_gaussian, exist_ok=True)
os.makedirs(output_median, exist_ok=True)

for filename in os.listdir(input_folder):
    if filename.endswith(".png"):
        path = os.path.join(input_folder, filename)

        image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

        gaussian = cv2.GaussianBlur(image, (5, 5), 0)

        median = cv2.medianBlur(image, 5)

        cv2.imwrite(os.path.join(output_gaussian, filename), gaussian)
        cv2.imwrite(os.path.join(output_median, filename), median)

print("Denoising completed for all frames.")
```
