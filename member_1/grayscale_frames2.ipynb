```python
import cv2
import os

os.makedirs("frames", exist_ok=True)

cap = cv2.VideoCapture("gray_video02.mp4")

count = 0
saved = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    if count % 10 == 0:
        cv2.imwrite(f"frames/frame_{saved}.png", frame)
        saved += 1

    count += 1

cap.release()

print("Frames extracted successfully!")
```
